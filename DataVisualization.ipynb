```python
# 📦 Install required libraries (if not already)
!pip install matplotlib seaborn pandas scikit-learn

# 📚 Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

# 📂 Load your dataset (upload CSV if needed)
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))

# 👁️ Quick preview
df.head()

# 🎯 Target column
target_col = 'Role'



# 📈 Histograms of all numeric features
df.hist(figsize=(16, 12), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=18)
plt.tight_layout()
plt.show()

# 🔥 Correlation Heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df.drop(columns=[target_col]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# 📌 Pairplot of selected features
selected_features = ['Programming Skills', 'Communication skills', 'Openness', 'Emotional_Range', 'Project Management']
sns.pairplot(df[selected_features + [target_col]], hue=target_col, height=2.5)
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()

# 🧠 Optional: PCA 2D plot for career roles
features = df.drop(columns=[target_col])
labels = LabelEncoder().fit_transform(df[target_col])

pca = PCA(n_components=2)
components = pca.fit_transform(features)

plt.figure(figsize=(10, 7))
scatter = plt.scatter(components[:, 0], components[:, 1], c=labels, cmap='tab20', alpha=0.8)
plt.colorbar(scatter, label='Encoded Career Roles')
plt.title("PCA Projection of Career Profiles")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

```python
# Install necessary libraries (Colab only)
!pip install pandas seaborn matplotlib scikit-learn

# Import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# Upload your dataset (choose your CSV file)
uploaded = files.upload()

# Load the CSV into a DataFrame
df = pd.read_csv(next(iter(uploaded)))
df.head()
```

```python
df.hist(figsize=(16, 12), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=18)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))
corr = df.drop(columns=['Role']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
selected_features = ['Programming Skills', 'Communication skills', 'Openness', 'Project Management', 'Emotional_Range']
sns.pairplot(df[selected_features + ['Role']], hue='Role', height=2.5)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

X = df.drop(columns=['Role'])
y = LabelEncoder().fit_transform(df['Role'])

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab20', alpha=0.8)
plt.colorbar(scatter, label='Encoded Career Role')
plt.title("PCA Projection of Career Profiles")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```
